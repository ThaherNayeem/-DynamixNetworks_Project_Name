```python
##  Exploratory Data Analysis (EDA)
```

```python
# Movies Vs TV Shows
df['type'].value_counts().plot(kind = 'bar')
plt.title('Movies Vs TV Shows')
plt.show()
```

```python
# Top 10 countries
df['country'].value_counts().head(10).plot(kind = 'bar')
plt.title('Top 10 countries')
plt.show()
```

```python
# Ratings Distribution
df['rating'].value_counts().plot(kind = 'bar')
plt.xlabel('Rating')
plt.title('Rating Distribution')
plt.show()
```

```python
# Content Added Over Years
df['release_year'] = df['date_added'].dt.year

df['release_year'].value_counts().sort_index().plot(kind = 'line')
plt.title('Content added over Time')
plt.show()
```

```python
# Top Genres
df['listed_in'].value_counts().head(10).plot(kind = 'bar')
plt.title('Top Genres')
plt.show()
```

```python
## Visualizations
```

```python
# Pie Chart  Movies vs TV Shows
df['type'].value_counts().plot(kind = 'pie',autopct = '%1.1f%%')
plt.title('Movies Vs TV Shows')
plt.ylabel('')
plt.show()
```

```python
# Heatmap  Missing Values
plt.figure(figsize=(10,6))
sns.heatmap(df.isnull(), cbar=False)
plt.title("Missing Values Heatmap")
plt.show()
```

```python
# Heatmap  Correlation (Numerical)
df['release_year'] = df['date_added'].dt.year

sns.heatmap(df.corr(numeric_only = True ),annot = True)
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Multiple charts toghether 
plt.figure(figsize=(14,10))

# Chart 1
plt.subplot(2,2,1)
df['type'].value_counts().plot(kind='bar')
plt.title("Movies vs TV Shows")

# Chart 2
plt.subplot(2,2,2)
df['country'].value_counts().head(5).plot(kind='bar')
plt.xticks(rotation = 30)
plt.title("Top Countries")

# Chart 3
plt.subplot(2,2,3)
df['rating'].value_counts().head(5).plot(kind='bar')
plt.title("Top Ratings")

#Chart 4
plt.subplot(2,2,4)
df['release_year'] = df['date_added'].dt.year

df['release_year'].value_counts().sort_index().plot(kind = 'line')
plt.title('Content added over years')
plt.show()

plt.tight_layout()
plt.show()
```

```python
## Advanced Analysis
```

```python
# Content Growth Trend by Type
df['release_year'] = df['date_added'].dt.year

growth = df.groupby(['release_year','type']).size().unstack()

growth.plot(kind = 'line', figsize = (10,6))
plt.title('Netflix Growth: Movies vs TV Shows')
plt.xlabel('Count')
plt.show()
```

```python
# Country Contribution Over Time
top_countries = df['country'].value_counts().head(5).index

country_trend = df[df['country'].isin(top_countries)]
country_trend = country_trend.groupby(['release_year','country']).size().unstack()

country_trend.plot(figsize = (10,6))
plt.title('Top Countries Content Trend')
plt.show()
```

```python
# Genre Popularity Analysis
geners = df['listed_in'].str.split(',', expand = True).stack()
geners = geners.str.strip()

geners.value_counts().head(10).plot(kind = 'bar',color = ['red','blue','green','purple','orange','cyan','magenta','yellow','brown','pink'])
plt.title('Top 10 Genres on Netflix')
plt.show()
```

```python
# Rating vs Content Type
pd.crosstab(df['rating'], df['type']).plot(kind = 'bar', figsize = (10,6))
plt.title('Rating VS Type')
plt.xlabel('Rating')
plt.show()
```

```python
# Director Productivity
df['director'].value_counts().head(10).plot(kind = 'bar',color = ['red','blue','green','purple','orange','cyan','magenta','yellow','brown','pink'])
plt.title('Top Directors on Netflix')
plt.show()
```

```python
# Content Release Year vs Added Year
plt.scatter(df['release_year'], df['date_added'])
plt.title('Released Year VS Date Added')
plt.xlabel('Released Year')
plt.ylabel('date_added')
plt.show()
```
